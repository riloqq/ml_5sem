4. Как найти самое часто встречаемое значение между двумя массивами? 

````python
import numpy as np

arr1 = np.array([1, 2, 2, 3, 4, 4, 4])
arr2 = np.array([2, 1, 3, 2, 1, 1, 4])

combined = np.concatenate([arr1, arr2])
unique_values, counts = np.unique(combined, return_counts=True)
most_common_value = unique_values[np.argmax(counts)]

print(most_common_value)  # Вывод: 1
````

8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
cartesian = np.random.uniform(-10, 10, (10, 2))
polar = np.column_stack((
    np.sqrt(np.sum(cartesian**2, axis=1)),  # радиусы
    np.arctan2(cartesian[:, 1], cartesian[:, 0])  # углы
))

print("Декартовы -> Полярные:")
for i in range(10):
    x, y = cartesian[i]
    r, phi = polar[i]
    print(f"({x:6.2f}, {y:6.2f}) -> (r ={r:6.2f}, φ ={phi:6.2f} рад)")
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
sleep_25 = df['Sleep_Hours'].quantile(0.25)
sleep_75 = df['Sleep_Hours'].quantile(0.75)

well_rested = df[df['Sleep_Hours'] >= sleep_75]  
sleep_deprived = df[df['Sleep_Hours'] <= sleep_25]  

well_rested_avg = well_rested['Exam_Score'].mean()
sleep_deprived_avg = sleep_deprived['Exam_Score'].mean()

print(f"Средний балл высыпающихся студентов: {well_rested_avg:.2f}")
print(f"Средний балл недосыпающих студентов: {sleep_deprived_avg:.2f}")
print(f"Высыпающиеся сдают лучше: {well_rested_avg > sleep_deprived_avg}")
````

6. Влияет ли физическая активность на результаты экзамена?

````python
high_activity = df[df['Physical_Activity'] >= 3]['Exam_Score'].mean()
low_activity = df[df['Physical_Activity'] < 3]['Exam_Score'].mean()
print(f"Средний балл активных студентов: {high_activity:.2f}")
print(f"Средний балл неактивных студентов: {low_activity:.2f}")
print(f"Активные студенты сдают лучше: {high_activity > low_activity}")
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
higher_education = df[df['Parental_Education_Level'].isin(['College', 'Postgraduate'])]['Exam_Score'].mean()

others = df[~df['Parental_Education_Level'].isin(['College', 'Postgraduate'])]['Exam_Score'].mean()

print(f"Средний балл студентов с родителями с высшим образованием: {higher_education:.2f}")
print(f"Средний балл остальных студентов: {others:.2f}")
print(f"Студенты с образованными родителями сдают лучше: {higher_education > others}")
````
